```python
import pytesseract
from pytesseract import Output
from PIL import Image
import cv2
img = cv2.imread("imgs/dilated_image1.jpg")

cv2_imshow(img)
cv2.waitKey(0)

d = pytesseract.image_to_data(img, output_type=Output.DICT)
print(d.keys())
```

```python
img_object = Image.open("imgs/dilated_image1.jpg")
img_text = pytesseract.image_to_string(img_object)
split_img = img_text.split('\n')

#print(img_text)
for string in split_img:
  if string == " " or string == "":
    continue
  else:
    print(string)
    break

```
